# for loop

```python
degrees=[70,80,90,40,90,55,75,85]    #create a list
print(degrees)
```

```python
for i in degrees:                  #for loop, i  is a variable for each element in the list
  print(i)
```

```python
for i in degrees:
  if(i >= 90):
    print("A")
  elif(i >= 80):
    print("B")
  elif(i >= 70):
    print("C")
  elif(i >= 60):
    print("D")
  else:
    print("F")

print("done")
```

```python
#for loop with break  (break stops the loop)
for i in degrees:

  if(i >= 90):
    print("A")
  elif(i >= 80):
    print("B")
    break
  elif(i >= 70):
    print("C")
  elif(i >= 60):
    print("D")
  else:
    print("F")

print("done")
```

```python
#for loop with continue (cotinue skips the current iteration and moves to the next one)
for i in degrees:
  if(i >= 90):
    print("A")
  elif(i >= 80):
    print("B")
  elif(i >= 70):
    print("C")
  elif(i >= 60):
    print("D")
  else:
    continue
    print("F")

print("done")
```

# while loop

```python
#while loop (infinite loop because the condition is always true)
number=0

while(number < 10):
  print(number)
```

```python
#while loop with increment so that the loop will run 10 times then stop
number=0

while(number<10):
   print(number)
   number+=1
```

```python
#while loop with break
while True:
  a = input("enter a one letter ")
  if(a == "z"):
    break
```
