#Continue the Pandas


```python
#import numpy
import numpy as np
```

```python
#create a new column with values by where
data['evaluation']=np.where(data['salary']>= 600,  'yes','no' )
data
```

```python
#DataFrame using the apply() method to create a new column named good in the DataFrame,
#apply() is used to apply the function to each row in the salary column.


def check_salary(salary):
    if salary >= 600:
        return 'yes'
    else:
        return 'no'
data['evaluation2'] = data['salary'].apply(check_salary)
data
```

```python
#map: Applying the function to create a new column
data['evaluation3'] = data['salary'].map(check_salary)
data
```

**fill :**


 -Fill with a Specific Value

 -Fill with the Mean, Median, or Mode

```python
data.isnull().sum()
```

```python
# Fill with 0
data['salary'].fillna(value=0, inplace=True)
```

```python
#fillna: used to replace all null values in a salary by a mean value.
data["salary"].fillna(data["salary"].mean(), inplace=True)
```

```python
#fillna: used to replace all null values in an experience by a median value.
data["Experience"].fillna(data["Experience"].median(),inplace=True)
```

```python
#fillna: used to replace all null values in a leaveornot by a mode value.
data["LeaveOrNot"].fillna(data["LeaveOrNot"].mode()[0], inplace=True)
```

```python
#duplicated: shows whether each row in the DataFrame is a duplicate or not.
data.duplicated()
```

```python
#drop_duplicates:removes duplicate rows from the DataFrame, keeping the first occurrence of each duplicated row by default.
data.drop_duplicates()
```
